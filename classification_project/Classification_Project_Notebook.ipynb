# Complete Visual Classification Manual (The Alarm System)
Welcome to the Complete Visual Manual for the **Critical Risk Alarm System**. 

While our regression models focus on predicting exact temperatures, this classification branch is designed purely for **safety and alarms**. We mathematically defined the top 10% hottest historical engine states as "Critical Risk" (1), and everything else as "Normal" (0). We then trained AI algorithms to act as a binary alarm system.

Below is the complete encyclopedia of all **31 diagnostic graphs** generated by our pipeline, along with plain-English explanations so you understand exactly how the AI evaluates risk.

## Chapter 1: The Alarm System Shootout
Because 90% of the time the engine is safe, a lazy AI could just guess "Safe" 100% of the time and still score 90% accuracy! 
To prevent this, we use a technique that mathematically punishes the AI severely if it fails to catch a critical alarm. We pitted 5 different algorithms against each other.

![Accuracy Comparison](classification_plots/01_accuracy_comparison.png)

### 💡 Graph Explanation: Classification Accuracy
**What is this?** This graph shows how often each AI correctly guessed if the engine was safe or failing.
**How to read it:** The higher the bar, the smarter the AI. Notice how the advanced boosting models (like LightGBM and XGBoost) score the highest, making them the most reliable alarm system for your engine!

## Chapter 2: The Core Diagnostics (Confusion Matrices)
Accuracy scores can sometimes be deceiving. The best way to evaluate an alarm system is a "Confusion Matrix", which breaks down exactly what kind of mistakes the AI made.

### The Master Matrix
![Confusion Matrix Master](classification_plots/02_confusion_matrix.png)

### 💡 Graph Explanation: The Confusion Matrix
**What is this?** A detailed grid of where the AI got it right, and where it got it wrong.
**How to read it:** 
- **Top-Left (Normal/Normal):** Engine was safe, AI correctly said it was safe. (Good!)
- **Bottom-Right (Critical/Critical):** Engine was failing, AI correctly triggered the alarm. (Good!)
- **Bottom-Left (Critical/Normal):** Engine was failing, but the AI missed it. (This is a dangerous "False Negative").
- **Top-Right (Normal/Critical):** Engine was safe, but the AI triggered a false alarm. (This is a safe "False Positive"). 

We ideally want all numbers to be clustered in the Top-Left and Bottom-Right boxes!

### Deep Dive: Train vs Test Consistency

## Chapter 5: Sensor-by-Sensor Diagnostics (Violin Plots)
Below are 12 "Violin Plots", one for each sensor. 

### 💡 Graph Explanation: How to read a Violin Plot
- The "bulge" (the thickest part of the violin) shows the most common, normal operating range for that sensor.
- The long, skinny "tails" show extreme anomalies. 
- We split the violins by color: **Blue (Normal)** vs **Orange (Critical Risk)**. By looking at where the Orange bulges compared to the Blue, you can instantly see the exact sensor threshold that triggers an engine failure!
